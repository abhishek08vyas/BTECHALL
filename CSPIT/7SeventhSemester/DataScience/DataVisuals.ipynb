## **Practical 5 : Anatomy of MatPlotLib chart image, Histogram, Pichart, Line Plot, Scatter Plot(Loan Prediction Dataset)** 

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fullData = pd.read_csv("LoanPrediction.csv")
#Imputing Missing values with mean for continuous variable
fullData['LoanAmount'].fillna(fullData['LoanAmount'].mean(), inplace=True)

fullData['Loan_Amount_Term'].fillna(fullData['Loan_Amount_Term'].mean(), inplace=True)
fullData['ApplicantIncome'].fillna(fullData['ApplicantIncome'].mean(), inplace=True)
fullData['CoapplicantIncome'].fillna(fullData['CoapplicantIncome'].mean(), inplace=True)

#Imputing Missing values with mode for categorical variables
fullData['Gender'].fillna(fullData['Gender'].mode()[0], inplace=True)
fullData['Married'].fillna(fullData['Married'].mode()[0], inplace=True)
fullData['Dependents'].fillna(fullData['Dependents'].mode()[0], inplace=True)
fullData['Loan_Amount_Term'].fillna(fullData['Loan_Amount_Term'].mode()[0], inplace=True)
fullData['Credit_History'].fillna(fullData['Credit_History'].mode()[0], inplace=True)
```

```python
plt.hist(fullData['ApplicantIncome'], 
         facecolor='peru', 
         edgecolor='blue', 
         bins=15)
plt.show()
```

```python
plt.hist(fullData['CoapplicantIncome'], 
         facecolor='orangered', 
         edgecolor='maroon', 
         bins=8)

plt.hist(fullData['ApplicantIncome'], 
         facecolor='peru', 
         edgecolor='blue', 
         bins=10, 
         alpha = 0.3)

plt.show()
```

```python
#boxPlot
fullData.boxplot(column='ApplicantIncome' ,by='Education')
```

```python
#Stackplot_Stemplot
y = np.vstack([fullData['Loan_Amount_Term'], 
               fullData['ApplicantIncome'], 
               fullData['CoapplicantIncome']])
labels = ['Loan_Amount_Term', 
          'ApplicantIncome', 
          'CoapplicantIncome']

# Colors for each stack
colors = ['sandybrown', 
          'tomato', 
          'skyblue']

#  Similar to pandas df.plot.area()
plt.stackplot(fullData['Gender'], y, 
              labels=labels,
              colors=colors, 
              edgecolor='black')

# Plots legend to the upperleft of Figure
plt.legend(loc=2)

plt.show()
```

```python
#Barplot
plt.figure(figsize=(7,5))
plt.bar(fullData['Loan_Status'],fullData['ApplicantIncome'])
plt.xlabel('LoanStatus')
plt.ylabel('ApplicantIncome')
plt.title('Bar Chart')
plt.show()
```

```python
#ScatterPlot
plt.scatter(fullData['Loan_Amount_Term'],fullData['ApplicantIncome'])
plt.title('Loan Amount Term vs Applicant Income ')
plt.xlabel('Loan Amount Term')
plt.ylabel('Applicant Income')
plt.show()
```

```python
print(pd.crosstab(fullData["Education"],fullData["Loan_Status"]))
Education = pd.crosstab(fullData["Education"],fullData["Loan_Status"])
Education.div(Education.sum(1).astype(float),axis=0).plot(kind="bar",stacked=True,figsize=(4,4))
plt.xlabel("Education")
plt.ylabel("Percentage")
plt.show()
```

```python
print(pd.crosstab(fullData['Dependents'],fullData["Loan_Status"]))
Dependents = pd.crosstab(fullData['Dependents'],fullData["Loan_Status"])
Dependents.div(Dependents.sum(1).astype(float),axis=0).plot(kind="bar",stacked=True,figsize=(4,4))
plt.xlabel("Dependents")
plt.ylabel("Percentage")
plt.show()
```
